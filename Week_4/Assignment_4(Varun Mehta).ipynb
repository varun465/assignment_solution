```python
import yfinance as yf
import pandas as pd



start_date = '2013-08-18'
end_date = '2023-08-18'


AAPL_data = yf.download('AAPL', start=start_date, end=end_date)['Close']
GOOGL_data = yf.download('GOOGL', start=start_date, end=end_date)['Close']
META_data = yf.download('META', start=start_date, end=end_date)['Close']

AAPL_data
GOOGL_data
META_data

```

```python
#2. merge all three dataframes on Date column
```

```python
z = pd.merge(AAPL_data, GOOGL_data, on='Date' , suffixes=('_AAPL', '_GOOGL'))
data1 = pd.merge(z, META_data, on = 'Date' , suffixes = ('_META'))
```

```python
new_column_names = {'Close_AAPL': 'AAPL', 'Close_GOOGL': 'GOOGL' , 'Close':'META' }
new_data = data1.rename(columns=new_column_names)
new_data
```

```python
#3. calculate log returns of the close price for each of the stock
```

```python
stock_return = np.log(new_data/new_data.shift(1))
stock_return
```

```python
# 4. Calculate 99 percentile, 95 percentile, 90 percentile VaR number for the portfolio
```

```python
# Define confidence levels for VaR calculation
confidence_levels = [0.99, 0.95, 0.90]

# Calculate VaR at different percentiles
var_results = {}
for confidence in confidence_levels:
    stock_return['PortfolioReturns'] = stock_return.sum(axis=1)
    var_results[confidence] = np.percentile(stock_return['PortfolioReturns'], 1 - confidence)

print("VaR results:")
for confidence, var in var_results.items():
    print(f"{int(confidence * 100)}th percentile VaR: {var:.2f}")
```

```python
# 5. Repeat the above exercise for different holding period such as 1day, 2day, 5day and 10day respectively
```

```python
#Define confidence levels for VaR calculation
confidence_levels = [0.99, 0.95, 0.90]

#Define holding periods (in days)
holding_periods = [1, 2, 5, 10]

#Calculate VaR using quantile for different holding periods and confidence levels
var_results = {}
for holding_period in holding_periods:
    var_results[holding_period] = {}
    for confidence in confidence_levels:
        # Calculate VaR using the quantile formula
        var_results[holding_period][confidence] = np.percentile(stock_return['PortfolioReturns'], 1 - confidence) * np.sqrt(holding_period)
 
 #Print VaR results
for holding_period, confidence_dict in var_results.items():
    print(f"VaR results for {holding_period}-day holding period:")
    for confidence, var in confidence_dict.items():
        print(f"{int(confidence * 100)}th percentile VaR: {var:.4f}")
    print()
```
